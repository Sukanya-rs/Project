# 1. Measuring the time taken to process the code. 



```python
import time
from time import process_time

start = process_time()

##code
def ascii_caesar_shift(message, distance):
    encrypted = ""
    for char in message:
      value = ord(char) + distance
      encrypted += chr(value % 128) #128 for ASCII
    return encrypted


##code

print(time.process_time() - start) 

ascii_caesar_shift(input(), 5)
```

# 3. With tic toc 

```python
import time 

tic = time.clock()

def ascii_caesar_shift(message, distance):
    encrypted = ""
    for char in message:
      value = ord(char) + distance
      encrypted += chr(value % 128) #128 for ASCII
    return encrypted

ascii_caesar_shift(input(), 5)

toc = time.clock()
print(toc - tic) 
```
